```python
import pandas as pd
import numpy as np

genres_df=pd.read_csv('tmdb_genres.csv')
genres_df
```

```python
movies_df=pd.read_csv('tmdb_movies.csv')
movies_df
```

# Szukanie wartości trzeciego kwartyla

```python
movies_df.describe()
```

```python
vote_third_quartile=6.8
```

# Baza Danych filmów z oceną większą bądź równą trzeciego kwartyla

```python
movies_df2=movies_df[movies_df['vote_average'].gt(vote_third_quartile)]
```

```python
movies_df2=movies_df2.sort_values(by='vote_average',ascending=False)
movies_df2=movies_df2.head(10)
movies_df2
```

# Lista top 10 filmów wg oceny widzów

```python
top10=movies_df2[['original_title','vote_average']]
top10=top10.reset_index(drop=True)
top10
```

# Baza danych filmów opublikowanych w latach 2010-2016:

```python
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'])
movies_df3=movies_df.sort_values(by='release_date', ascending=False)
movies_df4=movies_df3.loc[(movies_df3['release_date']>='2010-01-01') & (movies_df3['release_date'] <= '2016-12-31')]

movies_df4['year']=movies_df4['release_date'].dt.year.copy()
revenues=movies_df4.groupby('year')['revenue'].mean()
budgets=movies_df4.groupby('year')['budget'].mean()
revenues
budgets
```

```python
import matplotlib.pyplot as plt
```

```python
def million(x, pos):
        return '{:2.1f}M'.format(x*1e-6)
formatter = plt.FuncFormatter(million)
```

```python
plt.bar(revenues.index ,label='przychód', color='blue',height=revenues)
plt.plot(budgets, label='budżet', color='red')
plt.gca().yaxis.set_major_formatter(formatter)
plt.legend()
plt.title("Średni przychód i budżet filmu w latach 2010-2016")
plt.show()
```

# Łączę bazę danych filmów z nazwami gatunków filmów:

```python
genres_df = genres_df.rename(columns={'Unnamed: 0':'genre_id'})
genres_df
movies_df_genres=pd.merge(genres_df,movies_df, on='genre_id')
movies_df_genres['count']=1
movies_df_genres
```

```python
genres_count_df=movies_df_genres.pivot_table(values='count',index='genres',aggfunc=np.sum)
genres_count_df
```

# Odpowiedź: W bazie danych filmów znajduje się najwięcej dramatów (1207).

```python
genres_runtime_df=movies_df_genres.pivot_table(values='runtime',index='genres',aggfunc=np.mean)
genres_runtime_df
```
